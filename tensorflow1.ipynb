```python
# prompt: use the above to print different names

# Create a TensorFlow session
sess = tf.compat.v1.Session()

# Define the placeholder for the name
str_name = tf.compat.v1.placeholder(tf.string)

# Define the operation to concatenate "I am" with the name
name1 = "I am " + str_name

# Run the session with different names
print(sess.run(name1, feed_dict={str_name: "Alice"}))
print(sess.run(name1, feed_dict={str_name: "Bob"}))
print(sess.run(name1, feed_dict={str_name: "Charlie"}))

# Close the session
sess.close()
```

```python
# creating a placeholder
a = tf.placeholder(tf.float32, None)

# creating an operation
b = a + 10

# creating a session
with tf.Session() as session:

    # feeding data in the placeholder
    operation_res = session.run(b, feed_dict={a: [10, 20, 30, 40]})
    print("after executing the operation: " + str(operation_res))
```
