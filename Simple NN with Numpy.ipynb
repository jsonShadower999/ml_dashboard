```python
import numpy as np

# Features
x1 = 4
x2 = 5
x3 = 3

# Target variable
y = 100

# Learning rate
lr = 0.005

# Weights
w1 = 0
w2 = 0
w3 = 0
```

```python

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python

# Training the perceptron
for epoch in range(50):
    # Calculate the weighted sum
    weighted_sum = w1 * x1 + w2 * x2 + w3 * x3

    # Apply the sigmoid activation function
    y_pred = sigmoid(weighted_sum)

    # Compute the error
    error = (y - y_pred)**2

    # Compute the gradients
    dEw1 = 2 * (y - y_pred) * y_pred * (1 - y_pred) * x1         #-5
    dEw2 = 2 * (y - y_pred) * y_pred * (1 - y_pred) * x2         #+10
    dEw3 = 2 * (y - y_pred) * y_pred * (1 - y_pred) * x3

    # Update weights
    w1 = w1 - lr * dEw1      # 0-(-5X0.01).  =.   0+0.5 = 0.5
    w2 = w2 - lr * dEw2      #0-(10X0.1) =. 0-1 = -1
    w3 = w3 - lr * dEw3

    print("Epoch:", epoch, "Value of w1:", w1, "Value of w2:", w2, "Value of w3:", w3, "Error:", error, "Predicted:", y_pred)
```

```python
# prompt: use the above code and replace sigmoid with relu function

# Features
x1 = 2
x2 = 5
x3 = 3
# Target variable
y = 100

# Learning rate
lr = 0.005

# Weights
w1 = 0
w2 = 0
w3 = 0


# ReLU activation function
def relu(x):
    return max(0, x)


# Training the perceptron
for epoch in range(50):
    # Calculate the weighted sum
    weighted_sum = w1 * x1 + w2 * x2 + w3 * x3

    # Apply the ReLU activation function
    y_pred = relu(weighted_sum)

    # Compute the error
    error = (y - y_pred)**2

    # Compute the gradients
    dEw1 = 2 * (y - y_pred) * (x1 if y_pred > 0 else 0)
    dEw2 = 2 * (y - y_pred) * (x2 if y_pred > 0 else 0)
    dEw3 = 2 * (y - y_pred) * (x3 if y_pred > 0 else 0)

    # Update weights
    w1 = w1 - lr * dEw1
    w2 = w2 - lr * dEw2
    w3 = w3 - lr * dEw3

    print("Epoch:", epoch, "Value of w1:", w1, "Value of w2:", w2, "Value of w3:", w3, "Error:", error, "Predicted:", y_pred)
```
